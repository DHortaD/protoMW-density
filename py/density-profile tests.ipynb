```python
import numpy as np
import matplotlib.pyplot as plt
import tqdm
```

```python
r_test = np.linspace(0.01,30,128)

ns = np.linspace(1,10,10)
reb = np.linspace(1,50,50)

rs_ = np.zeros((10,50,128))

for indx, i in tqdm.tqdm_notebook(enumerate(ns)):
    for jndx, j in enumerate(reb):

        dn = 3.*i - 0.3333 + 0.0079/i

        rs_[indx, jndx] = (np.exp(-dn * ((r_test/j)**(1/i) -1)))
```

```python
reb
```

```python
colours = np.arange(500)+1
```

```python
sm = plt.cm.ScalarMappable(cmap=plt.cm.plasma_r, norm=plt.Normalize(vmin=1,\
                                                                    vmax=50))
norm = plt.Normalize()
col = plt.cm.plasma_r(norm(colours))
col = col.reshape(10,50,4)
```

```python
rs_.shape
```

```python
plt.figure(figsize=(10.5,8))
for indx, i in enumerate(rs_):
    for jndx, j in enumerate(i):
        plt.plot(j, label='$n$: '+str(ns[indx])+'; $r_{e}$'+str(reb[jndx]), c=col[indx][jndx])
plt.yscale('log')
plt.legend(loc='lower left', ncols=5, bbox_to_anchor=[0., 1])
```

```python
rs_.shape
```

```python
plt.figure(figsize=(16,6), facecolor='white')

plt.subplot(1,2,1)
plt.title('Varying $n$', fontsize=20)
for indx, i in enumerate(rs_):
    plt.plot(i[5], label='$n$: '+str(ns[indx]))
plt.yscale('log')
plt.legend(loc='lower left') 
plt.xlabel('$r$', fontsize=20)
plt.ylabel('Density', fontsize=20)

plt.subplot(1,2,2)
plt.title('Varying scale radius', fontsize=20)
for indx, i in enumerate(rs_[2]):
    plt.plot(i, label='$r_{e}$: '+str(reb[indx]))
plt.yscale('log')

plt.legend(loc='upper right', ncols=4, bbox_to_anchor=[0.8,2]) 
plt.xlabel('$r$', fontsize=20)
plt.ylabel('Density', fontsize=20)
```
